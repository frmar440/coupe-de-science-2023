# Onde radio

Votre antenne de télécommunications capte une onde radio d'amplitude et de fréquence inconnues. Pour décoder le message contenu dans cette onde, vous devez synchroniser votre décodeur avec la fréquence de l'onde. Pour rappel, la fréquence d'une onde sinusoïdale est l'inverse de la période d'un cycle:
$$f = \frac{1}{T},$$
où $T$ est la période (en secondes), c'est-à-dire la durée d'un cycle de la fonction sinusoïdale. L'interface de votre antenne de télécommunications permet l'affichage de l'onde captée:

```python
%matplotlib inline

import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()
ax.set_xlim(0, .6)
ax.set_ylim(-1.1, 1.1)
ax.set_xlabel("Temps [s]")
ax.set_ylabel("Amplitude [V/m]")
ax.grid()

t = np.linspace(0, .6, 100)
ax.plot(t, np.sin(10*np.pi*t))

plt.show()
```
